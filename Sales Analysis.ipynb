### Task 1: Merge the 12 months of sales data into a single CSV file

```python
files = [file for file in os.listdir('./Sales_Data')]
all_months_data = pd.DataFrame()

for file in files:
    df = pd.read_csv("./Sales_Data/"+file)
    all_months_data = pd.concat([all_months_data, df])
    
all_months_data.to_csv("all_data.csv", index=False)
```

## Q3. What time should we display advertisements to maximize likelihood of customer's buying product?

```python
all_data['Order Date'] = pd.to_datetime(all_data['Order Date'])
all_data
```

```python
all_data['Hour'] = all_data['Order Date'].dt.hour
all_data['Minute'] = all_data['Order Date'].dt.minute
all_data.head()
```

## Q4. What products are most often sold together?

```python
all_data.head()
```

```python
print(f"No of duplicate rows in column Order ID = {all_data['Order ID'].shape[0] - len(pd.unique(all_data['Order ID']))}")
```

```python
df = all_data[all_data['Order ID'].duplicated(keep=False)]

df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))

df = df[['Order ID', 'Grouped']].drop_duplicates()

df.head()
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))
     
# print(count)
# count.most_common(10)

for key, value in count.most_common(10):
    print(key, value)
```

## Q5. What product sold the most? Why do you think it sold the most?

```python
all_data.head()
```

```python
product_group = all_data.groupby('Product')
product_group.sum()
```

```python
print([product for product, df in product_group])
```

```python
quantity_ordered = product_group.sum()['Quantity Ordered']

products = [product for product, df in product_group]

plt.figure(figsize=(13, 8))
plt.bar(products, quantity_ordered)
plt.ylabel('Quantity Ordered')
plt.xlabel('Product')
plt.xticks(products, rotation='vertical', size=8)
plt.show()
```

```python
prices = all_data.groupby('Product').mean()['Price Each']
prices
```

```python
# plot secondary y-axis for prices
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(products, quantity_ordered, color='g')
ax2.plot(products, prices, 'b-')

ax1.set_xlabel('Product')
ax1.set_ylabel('Quantity Ordered', color='g')
ax2.set_ylabel('Price ($)', color='b')
ax1.set_xticklabels(products, rotation='vertical', size=10)

plt.show()
```

It shows that the price of the products with higher quantities is lower and vice-versa.
